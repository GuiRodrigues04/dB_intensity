```python
import pandas as pd
import numpy as np


# Reading excel of the data
Tabelas = []
for i in range(14):
    Tabelas.append(pd.read_excel(f'./{i + 1}.xls'))

    
# Transforming in dB
Tabelas_1 = []
for i in range(14):
    Tabelas_1.append(pd.read_excel(f'./{i + 1}.xls'))
for i in Tabelas_1:
    i['audio amplitude (a.u.)'] = 10 * (np.log10( i['audio amplitude (a.u.)'] / 2))
```
